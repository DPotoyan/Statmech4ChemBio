# Quiz: Mean Field, Landau, and Flory-Huggins

Test your understanding of the mean-field approximation for the Ising model, Landau theory of phase transitions, and Flory-Huggins theory of binary mixtures.


```python
import ipywidgets as widgets
from IPython.display import display, Markdown

def mc_question(prompt, options, correct_index):
    """Display a multiple-choice question with feedback."""
    out = widgets.Output()
    buttons = [widgets.Button(description=f"{chr(65+i)}", layout=widgets.Layout(width='60px')) for i in range(len(options))]

    display(Markdown(f"**{prompt}**"))
    for i, opt in enumerate(options):
        display(Markdown(f"&emsp;**{chr(65+i)}.** {opt}"))

    box = widgets.HBox(buttons + [out])
    display(box)

    def check(b):
        idx = buttons.index(b)
        out.clear_output()
        with out:
            if idx == correct_index:
                display(Markdown("&ensp; ✅ **Correct!**"))
            else:
                display(Markdown(f"&ensp; ❌ Try again."))
    for b in buttons:
        b.on_click(check)
```
