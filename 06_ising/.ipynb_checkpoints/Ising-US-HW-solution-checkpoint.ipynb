```python
# Make some convenience plotting/analysis functions 

def my_hist(m, W, bins=40):
    
    '''A simpe hist function for making/ploting umbrella sampling data '''
    
    hist, edges = np.histogram(m,bins, range=(np.min(W), np.max(W)),density=True)
    edges = 0.5*(edges[1:]+edges[:-1])
    
    return edges, hist

def my_us_run_and_plot(spins, W):
    
    '''Input: spin config and umbrella params.
    Runs US between minW maxW from spins state 
    Output: return final spin configuration and the histogram over m data'''

    fig, ((ax1,ax2),(ax3,ax4))  = plt.subplots(nrows=2,ncols=2,figsize=(8,6))
    
    
    ax1.imshow(spins) # state before run
    configs,   m   = us_mcmc(spins, J, B, T, W , n_steps = 1e6, out_freq = 1e5) # equil
    configs,   m   = us_mcmc(configs[-1], J, B, T, W , n_steps = 5e6, out_freq = 1e4) # production
    ax2.imshow(spins) # state after run
 
    X, P_X = my_hist(m, W) # Histogram
    
    ax3.plot(X, P_X, '-o') # Plot histogram
    ax4.plot(m)           # Plot mcmc trajectory
    
    print(['Min(m), ', np.min(m), 'Max(m) ', np.max(m), 'std ', np.std(m)])
    
    return configs[-1], X, P_X 
```

#### Tips and tricks of running US

- Run Umbrella sampling from first window to last window secuentially or parallel. But parlalle will be a little slower. 

- Initialize from a low energy state and study fluctuations of order parameter to determine size of the window!
