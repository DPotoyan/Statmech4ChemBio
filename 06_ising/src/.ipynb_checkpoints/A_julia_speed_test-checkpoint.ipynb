## Computing thermo observables

```julia
fig, ax = subplots(ncols=2, figsize=(10,4))

time = 1:5000:10^7

cut = 100:2000

ax[1].plot(time[cut], Es[cut], color = "red"  )

ax[2].plot(time[cut], Ms[cut], color = "green");
```

## Running multiple trajectories

```julia
Cv = []
Xi = []

temps = 1:0.2:5

@time for T in temps
    
    spins, Ms, Es = ising2d_mcmc(N, J, B, T)
    
    push!(Cv,      var(Es[1000:end])/T^2)
    push!(Xi,      var(Ms[1000:end])/T^2)
    
end
```

```julia
fig, ax = subplots(ncols=2, figsize=(10,4))

ax[1].plot(temps, Cv,"-o")
ax[2].plot(temps, Xi,"-o");
```
