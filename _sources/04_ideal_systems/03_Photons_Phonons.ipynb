### Solid of Einstein

Conisder a model of a solid with atoms repreented as $N$ localized and noninteracting quantim 1D oscillators with same fundmental frequency $\omega$. The solid is in thermal contact with a heat bath at temperature $T$.

The energy of this solid will be a sum of oscillator energies defined in terms of quantum numbers $n_1, n_2, ... n_N$

$$E(n_1, n_2, ...n_N) = \sum^{j=N}_{j=1}(n_j +1/2)\hbar \omega$$

Partition function can be decouples intp product of exponentials

$$Z = \sum_{n_1, n_2,...} exp \Big[- \sum^{j=N}_{j=1}(n_j +1/2)\beta \hbar \omega \Big] = \Bigg ( \sum^{n_1=\infty}_{n_1=0} exp \Big[-(n_1 +1/2)\beta \hbar \omega\Big ] \Bigg)^N = z^N_1$$

**partition function per oscillator**

$$Z_1 = \sum^{n_1=\infty}_{n_1=0} exp \Big[-(n_1 +1/2)\beta \hbar \omega \Big ] = \frac{exp(-\frac{1}{2}\beta \hbar \omega)}{1-exp(- \beta \hbar \omega)}$$

**free energy per oscillator**
$$f = -\beta^{-1} log Z_1 = \frac{1}{2}\hbar \omega +\beta^{-1} log\Big[1- exp(-\beta\hbar \omega) \Big]$$

**entropy per oscillator**

$$s= - \frac{\partial f}{\partial T}= -k_B log\Big[1- exp(-\beta\hbar \omega) \Big] +k_B (\beta \hbar\omega) \frac{exp(-\beta \hbar \omega)}{[1-exp(\beta \hbar \omega)]}$$

**heat capacity per oscillator**

$$c_v = T\frac{\partial s}{\partial T} = k_B (\beta \hbar \omega)^2 \frac{exp(\beta \hbar \omega)}{[1-exp(\beta \hbar \omega)]^2}$$

**average energy per oscillator**

$$u = \frac{\partial logZ_1}{\partial (-\beta)} = \frac{1}{2}\hbar \omega + \frac{\hbar \omega}{[exp(\beta \hbar \omega)-1]}$$

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
def c_v(x):
    '''x=beta*hbar*omega'''
    
    return x**2 * np.exp(x) / (1-np.exp(x))**2  

xs = np.linspace(0.1, 2, 1000)

plt.plot(1/xs, c_v(xs), lw=3)

plt.ylabel('$C_v$',fontsize=20)
plt.xlabel('$T$',fontsize=20)
```
